Import Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Line Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph
plt.figure(figsize=(5,3), dpi=300)

# 'b^--' => Shorthand Notation to Restyle Line
plt.plot(x,y, 'b^--',label='2x')

plt.title('Our First Graph!', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10,12])

## Line Number Two
x2 = np.arange(0,4,0.5)

plt.plot(x2[:6], x2[:6]**2, 'r',label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--',label='x^2')

plt.legend()
plt.show()
```

### Basic Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

plt.bar(labels, values)

plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)


patterns = ['/','*','.']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

### Histogram Example

```python
fifa = pd.read_csv('../dataset/fifa_data.csv')
fifa.head()
```

#### Task: How many player overal skills is between x,y? 

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.title('Distribution of Player Overall Skills in FIFA 2018')

plt.xlabel('Overal Skill Level')
plt.ylabel('Number of Players')
plt.xticks(bins)

plt.show()
```

### Pie Charts

Prefered foot

```python
fifa['Preferred Foot'].head()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
print(left)
```

```python
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
print(right)
```

```python
labels = ['Left', 'Right']
colors = ['#abcdef', '#aafffa']

plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

#### More Advanced Pie Charts
Range of Weight

```python
fifa.Weight.head()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight.head()
```

Value Seperate

```python
light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium = fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)].count()[0]
medium = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 175)].count()[0]
medium_heavy = fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200)].count()[0]
heavy = fifa.loc[fifa['Weight'] >= 200].count()[0]
```

```python
# It is just for Color
plt.style.use('ggplot')

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200','Over 200']

explode = (.4,.2,0,0,.4)

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.title('Weights Distribution of FIFA Players (in lbs)')

plt.show()
```

### Basic BoxPlot

Compare Team Players

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']

barcelona.head()
```

```python
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

madrid.head()
```

Boxplot

```python
labels=['Barcelona', 'Real Madrid']

plt.figure(figsize=(5,8))

plt.boxplot([barcelona, madrid], labels=labels)

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
